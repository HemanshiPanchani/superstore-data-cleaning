```python
# Fix Date Format
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d-%m-%Y')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], format='%d-%m-%Y')

print("✅ Dates Fixed!")
print(df[['Order Date', 'Ship Date']].head())
```

```python
# Flag Outliers (we keep them but mark them)
for col in ['Sales', 'Quantity', 'Discount', 'Profit']:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df[f'{col}_Outlier_Flag'] = ~df[col].between(lower, upper)

print("✅ Outliers Flagged!")
print(df[['Sales', 'Sales_Outlier_Flag', 'Profit', 'Profit_Outlier_Flag']].head())
```

```python
# Feature Engineering — Create New Useful Columns

# 1. Shipping Days (how long delivery took)
df['Shipping Days'] = (df['Ship Date'] - df['Order Date']).dt.days

# 2. Order Month, Year, Quarter
df['Order Month'] = df['Order Date'].dt.month
df['Order Year'] = df['Order Date'].dt.year
df['Order Quarter'] = df['Order Date'].dt.quarter

# 3. Profit Margin %
df['Profit Margin %'] = ((df['Profit'] / df['Sales']) * 100).round(2)

# 4. Revenue Category
df['Revenue Category'] = pd.cut(df['Sales'],
                                 bins=[0, 100, 500, 1000, float('inf')],
                                 labels=['Low', 'Medium', 'High', 'Premium'])

# 5. Is Discounted
df['Is Discounted'] = df['Discount'] > 0

print("✅ Feature Engineering Done!")
print(df[['Shipping Days', 'Order Month', 'Order Year', 
          'Profit Margin %', 'Revenue Category', 'Is Discounted']].head())
```

```python
# Save cleaned dataset
df.to_csv('cleaned_superstore.csv', index=False)
print("✅ Cleaned dataset saved as cleaned_superstore.csv!")
print(f"Final Shape: {df.shape}")
print(f"New columns added: {df.shape[1] - 21}")
```
